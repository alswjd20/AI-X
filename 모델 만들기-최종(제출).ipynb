# 모델 만들기

## cat boost 모델로 해보자

```python
from catboost import CatBoostRegressor
```

```python
cb = CatBoostRegressor(depth=6, iterations=900, learning_rate=0.07)
```

```python
cb.fit(train_x, train_y)
cb_pred = cb.predict(test_x)
```

```python
y_pred_cb =  pd.DataFrame({'id':range(0,1416), 'count':cb_pred})
y_pred_cb
```
